## Conditonal Column

```python
 avg_price = df['AveragePrice'].mean()
```

```python
avg_price
```

```python
# creating a conditional column (Price Category) that categorizes each price
df['price_cat'] = np.where(df['AveragePrice'] > avg_price, 'expensive',  'inexpensive')
```
